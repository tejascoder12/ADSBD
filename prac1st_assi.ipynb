````python
# Iterate over each column
for column in df.columns:
    # Check the data type of the column
    column_dtype = df[column].dtype
    
    # Determine if the column is categorical
    is_categorical = pd.api.types.is_categorical_dtype(column_dtype)
    
    # Print the result for each column
    if is_categorical:
        print(f"The column '{column}' is categorical.")
    else:
        print(f"The column '{column}' is not categorical.")
````

```

```
**The pd.factorize() function assigns a unique numeric code to each unique category in the specified column**. It replaces the categorical values with corresponding numeric codes. The result is a quantitative variable where each category is represented by a numeric value.

Choose the approach based on your specific requirements and the nature of the categorical variable in your DataFrame.** pd.get_dummies() is useful when you want to create binary columns for each category**, while **pd.factorize() is suitable when you want to assign numeric codes to each category.**
