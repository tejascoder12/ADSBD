```python
rscore=df['reading_score']
q1=np.percentile(rscore, 25)
q3=np.percentile(rscore, 75)
```

```python
IQR=q3-q1
IQR
```

```python
l_bound= q1-(1.5*IQR)
u_bound= q3+(1.5*IQR)
print(l_bound, u_bound)
```

```python
r_ouliers=[]
for i in rscore:
    if(i<l_bound or i>u_bound):
        r_outliers.append(i)
        print(r_outliers)
```
